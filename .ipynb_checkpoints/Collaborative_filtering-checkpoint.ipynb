# Système de recommandation de films 
* Modèle de recommandation basé sur le filtrage collaboratif 
* Méthode "matrix factorization" décrite dans le talk de Alexandros Karatzoglou : http://www.computervisiontalks.com/machine-learning-for-recommender-systems-with-alexandros-karatzoglou/
* Dataset : MovieLens (100k) disponible ici : http://grouplens.org/datasets/movielens/
  * 100,000 notes de 943 utilisateurs réparties parmi 1682 films 

```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import time
%matplotlib inline
```

## Training set utilisé pour l'optimisation des hyper-paramètres (80% de ml-100k)

```python
n_users = 943
n_movies = 1682

# Tableau répertoriant les notes données par les utilisateurs : user_id / movie_id / rating / timestamp
ratings = np.genfromtxt("ml-100k/u1.base", delimiter="	", dtype=np.int)
n_ratings = ratings.shape[0]

# Y(i,j) : note (de 1 à 5) donnée par l'utilisateur i au film j (0 par défaut si pas de note)
# R(i,j) : 1 si l'utilisateur i a noté le film j (0 sinon)
Y = np.zeros((n_users,n_movies),np.float64)
R = np.zeros((n_users,n_movies))

for i in range(ratings.shape[0]):
    Y[ratings[i][0]-1][ratings[i][1]-1] = ratings[i][2]
    R[ratings[i][0]-1][ratings[i][1]-1] = 1
print "Taille du training set : " + repr(n_ratings)


# Normalisation des notes : Y_norm(n_users,n_movies) et Y_mean(n_movies)
# Y_norm(i,j) = déviation de la note donnée par l'utilisateur i au film j, par rapport à la note moyenne donnée au film j
# (reste 0 si pas de note --> on prédira la valeur moyenne)
# Y_mean(j) = note moyenne donnée au film j par l'ensemble des utilisateurs
Y_norm = np.zeros(Y.shape)
Y_mean = np.zeros(Y.shape[1])

for j in range(Y.shape[1]):
    if Y[np.nonzero(Y[:,j]),j].size==0: # Films non notés dans le training set, on met leur note moyenne à zéro
        # Remarque : les critères x de ces films seront à 0 après entrainement, car seule leur régularisation joue dans J
        # --> On prédira toujours une note 0 pour ces films, quel que soit l'utilisateur
        # (ceci revient à ne pas recommander un film qui n'a jamais été noté)
        Y_mean[j] = 0
    else :
        Y_mean[j] = np.mean(Y[np.nonzero(Y[:,j]),j])
        Y_norm[np.nonzero(Y[:,j]),j] = Y[np.nonzero(Y[:,j]),j] - Y_mean[j]
```

## Test set utilisé pour l'optimisation des hyper-paramètres (20% de ml-100k)

```python
n_users = 943
n_movies = 1682

# Tableau répertoriant les notes données par les utilisateurs : user_id / movie_id / rating / timestamp
ratings_test = np.genfromtxt("ml-100k/u1.test", delimiter="	", dtype=np.int)
n_ratings_test = ratings_test.shape[0]

# Y_test(i,j) : note (de 1 à 5) donnée par l'utilisateur i au film j (0 par défaut si pas de note)
# R_test(i,j) : 1 si l'utilisateur i a noté le film j (0 sinon)
Y_test = np.zeros((n_users,n_movies),np.float64)
R_test = np.zeros((n_users,n_movies))

for i in range(ratings_test.shape[0]):
    Y_test[ratings_test[i][0]-1][ratings_test[i][1]-1] = ratings_test[i][2]
    R_test[ratings_test[i][0]-1][ratings_test[i][1]-1] = 1
print "Taille du test set : " + repr(n_ratings_test)


# Normalisation des notes : Y_test_norm(n_users,n_movies) et Y_test_mean(n_movies)
# Y_test_norm(i,j) = déviation de la note donnée par l'utilisateur i au film j, par rapport à la note moyenne donnée au film j
# (reste 0 si pas de note --> on prédira la valeur moyenne)
# Y_test_mean(j) = note moyenne donnée au film j par l'ensemble des utilisateurs
Y_test_norm = np.zeros(Y_test.shape)
Y_test_mean = np.zeros(Y_test.shape[1])
for j in range(Y_test.shape[1]):
    if Y_test[np.nonzero(Y_test[:,j]),j].size==0: # Films non notés dans le test set, on met leur note moyenne à zéro
        Y_test_mean[j] = 0
    else :
        Y_test_mean[j] = np.mean(Y_test[np.nonzero(Y_test[:,j]),j])
        Y_test_norm[np.nonzero(Y_test[:,j]),j] = Y_test[np.nonzero(Y_test[:,j]),j] - Y_test_mean[j]
```

## Définition d'une fonction pour calculer la fonction de coût : 
* Retourne : Le coût tenant compte de la régularisation
* Arguments :
  * y : array (n_users x n_movies) tel que y[i,j] = note attribuée au film j par l'utilisateur i
  * r : array (n_users x n_movies) tel que r[i,j] = 1 si l'utilisateur i a noté le film j
  * theta : array (n_users x n_features) des critères reflétant les "goûts" des utilisateurs
  * x : array (n_movies x n_features) des critères reflétant le "type" de films
  * rg : Paramètre de régularisation (pour simplifier, on utilise le même paramètre pour régulariser theta et x)

## Définition d'une fonction pour la minimisation par méthode du gradient :
* Retourne : [x,theta,iter_n,j_conv]
  * x : array (n_movies x n_features)
  * theta : array (n_user x n_features)
  * iter_n : nombre d'itération nécessaires à la minimisation
  * j_conv : valeur du coût (tenant compte de la régularisation) à la convergence
* Arguments :
  * y : array (n_users x n_movies) tel que y[i,j] = note attribuée au film j par l'utilisateur i
  * r : array (n_users x n_movies) tel que r[i,j] = 1 si l'utilisateur i a noté le film j
  * theta : array (n_users x n_features) des critères reflétant les "goûts" des utilisateurs (à déterminer pour minimiser J)
  * x : array (n_movies x n_features) des critères reflétant le "type" de films (à déterminer pour minimiser J)
  * rg : Paramètre de régularisation
  * al : "learning rate"
  * eps : critère de convergence
  * max_it : nombre maximal d'itérations pour la méthode du gradient

```python
# Fonction de coût
def Cost(y, r, theta, x, rg) :
    n_r = r[np.nonzero(r)].size
    error = (0.5/n_r)*np.einsum('ij,ji->', r*(np.dot(theta,x.T)-y), (r*(np.dot(theta,x.T)-y)).T)
    return error + (0.5*rg/n_r)*np.einsum('ij,ji->', theta, theta.T) + (0.5*rg/n_r)*np.einsum('ij,ji->', x, x.T)

# Méthode du gradient
def GradientDescent(y, r, theta, x, al, rg, eps, max_it):
    n_r = r[np.nonzero(r)].size
    j = Cost(y, r, theta, x, rg)
    j_hist = [j]
    j_old = j + 10*eps
    iter_n=0
    while abs(j-j_old) > eps :
        j_old = j
        Grad_X = (1.0/n_r)*np.dot((r*(np.dot(theta,x.T)-y)).T,theta) + (rg/n_r)*x
        Grad_Theta = (1.0/n_r)*np.dot(r*(np.dot(theta,x.T)-y),x) + (rg/n_r)*theta
        x = x - alpha*Grad_X
        theta = theta - alpha*Grad_Theta
        j = Cost(y, r, theta, x, rg)
        j_hist.append(j)
        iter_n=iter_n+1
        if iter_n==max_it :
            print "max_iter at limit"
            break
    plt.plot(j_hist)
    plt.ylabel("Fonction de cout")
    plt.xlabel("nb iterations")
    plt.show()
    print "n_iter = " + repr(iter_n)
    print "coût à la convergence :" + repr(j_hist[-1])
    return [x,theta,iter_n,j_hist[-1]]
```

## Choix du meilleur "learning rate" pour la méthode du gradient :
* On fixe arbitrairement le nombre de critères par utilisateur et par film : n_features = 20
* On ne s'occupe pas de la régularisation : reg = 0.0 (i.e. pas de régularisation)
* On réalise une minimisation pour différentes valeurs du "learning rate"

```python
# Nombre de critères
n_features = 20
# Paramètre de régularisation
reg = 0.0
# Condition de convergence
epsilon = 0.001
# Nombre maximal d'itération
max_iter = 100

alpha_values = [100.0,200.0,300.0,400.0,500.0,600.0,700.0,800.0]
iter_nb = []
J_last = []
for alpha in alpha_values :
    X = np.random.randn(n_movies,n_features)
    Theta = np.random.randn(n_users,n_features)
    print "alpha = " + repr(alpha)
    [X, Theta, a, b] = GradientDescent(Y_norm, R, Theta, X, alpha, reg, epsilon, max_iter)
    iter_nb.append(a)
    J_last.append(b)
    print "-------------------------"
```

* On affiche :
  * Le nombre d'itérations pour la minimisation en fonction de la valeur du "learning rate" alpha
  * Le coût à la convergence en fonction du "learning rate" alpha

```python
# Nombre d'itérations pour la minimisation, et cout à la convergence en fonction de alpha
plt.plot(alpha_values,iter_nb)
plt.ylabel("nb iterations")
plt.xlabel("alpha")
plt.show()

plt.plot(alpha_values,J_last)
plt.ylabel("cout a la convergence")
plt.xlabel("alpha")
plt.show()
```

Plusieurs remarques à propos du choix du "learning rate" : 
* Pour alpha >= 500, la méthode du gradient ne converge plus <-- Le pas est beaucoup trop grand
* alpha = 400 semble être la meilleure valeur car elle minimise le coût à la convergence ainsi que le nombre d'itérations
* On retient comme valeur alpha = 300 car en réalité alpha = 400 est "instable" (parfois la méthode du gradient ne converge pas avec alpha=400, en fonction des valeurs initiales de X et Theta)

_**Valeur retenue du "learning rate" pour la suite : alpha = 300**_

## Choix du paramètre de régularisation : 
* On cherche a déterminer la valeur du paramètre de régularisation reg, afin d'éviter un "overfitting" du training set
* Pour cela, on fait varier reg, et pour chaque valeur de reg :
  * On détermine les X et Theta qui minimisent la fonction de coût du training set
  * On réutilise ces X et Theta pour évaluer la fonction de coût sur le test set

```python
# Nombre de critères
n_features = 40
# Learning rate
alpha = 300.0
# Condition de convergence
epsilon = 0.001
# Nombre maximal d'itération
max_iter = 200

# Paramètre de régularisation
reg_values = [2.0,5.0,10.0,15.0,20.0,25.0,30.0,35.0,40.0]
iter_nb = []
J_train = []
J_test = []

for reg in reg_values :
    X = np.random.randn(n_movies,n_features)
    Theta = np.random.randn(n_users,n_features)
    print "reg = " + repr(reg)
    [X, Theta, a, b] = GradientDescent(Y_norm, R, Theta, X, alpha, reg, epsilon, max_iter) # Entrainement training set avec régularisation
    iter_nb.append(a)
    J_train.append(Cost(Y_norm, R, Theta, X, 0.0)) # Cout (sans les paramètres de régularisation)
    J_test.append(Cost(Y_test_norm, R_test, Theta, X, 0.0)) # Evaluation du coût sur le test set (sans régularisation)
    print "-------------------------"
```

* On affiche le coût sur le training set J_train (en bleu) et le coût sur le test set J_test (en rouge) en fonction du paramètre de régularisation :

```python
plt.plot(reg_values, J_train, "b", reg_values, J_test, "r")
plt.ylabel("cout")
plt.xlabel("parametre de regularisation")
plt.show()

print "regularisation : " + repr(reg_values)
print "-------------------------------------"
print "cout sur le train set : " + repr(J_train)
print "-------------------------------------"
print "cout sur le test set : " + repr(J_test)
```

* On retient comme valeur pour le paramètre de régularisation : reg = 15.0 (valeur qui minimise J_test)

_** Valeurs retenues pour la suite : **_
* _** alpha = 300 **_
* _** n_features = 40 **_
* _** reg = 15.0 **_

## Utilisation du modèle pour faire des recommandations pour un nouvel utilisateur :
* On utilise la matrice Y2 contenant les notes du training set précédent, plus celles données par le nouvel utilisateur
* On utilise la matrice R2 : R[i,j] = 1 si l'utilisateur i a noté le film j
* L'entrainement est fait sur les notes déjà données par l'ensemble des utilisateur + le nouvel utilisateur
* La prédiction est faite pour tous les utilisateurs et tous les films
* Pour les recommandations du nouvel utilisateur :
  * On réordonne les indices j correspondant au films selon l'ordre des notes prédites pour le nouvel utilisateur
  * On "jette" les films qu'il a déjà notés (on ne veut pas lui recommander un film qu'il a déjà vu)
  * On récupère les titres des films correspondant aux indices j des films (fichier u.item)
  * On affiche les titres des 20 films pour lesquels la note prédite est la plus grande
* Pour comparaison, on affiche les recommandations qui seraient faites uniquement sur la base de la note moyenne donnée par tous les utilisateurs à chaque film

```python
# Entrainement

# Nombre de critères
n_features = 40
# Paramètre de régularisation
reg = 15.0
# Learning rate
alpha = 300.0
# Condition de convergence
epsilon = 0.001
# Nombre maximal d'itération
max_iter = 200

X = np.random.randn(n_movies2,n_features)
Theta = np.random.randn(n_users2,n_features)
[X, Theta, a, b] = GradientDescent(Y2_norm, R2, Theta, X, alpha, reg, epsilon, max_iter)

# Prédictions (notes normalisées)
Y2_pred = np.dot(Theta,X.T)
# Prédictions (notes de 1 à 5)
for i in range(Y2_pred.shape[0]) :
    Y2_pred[i,:] = Y2_pred[i,:] + Y2_mean

# Recommandations pour le nouvel utilisateur

# Ordonne les films par ordre croissant de notes prédites
Movies_ord =  Y2_pred[-1,:].argsort() # indices : commencent à 0
Rates_ord = Y2_pred[-1,Y2_pred[-1,:].argsort()]
Already_viewed = R2[-1,Y2_pred[-1,:].argsort()]

# On jette les films qu'il a déjà vus
Movies_ord = Movies_ord[Already_viewed==0]
Rates_ord = Rates_ord[Already_viewed==0]

# Correspondance movie_id / Titre du film
Title = [] # 'Toy Story' (movie_id=1) sera à l'indice 0
with open('ml-100k/u.item') as csvfile:
    reader = csv.reader(csvfile, delimiter='|')
    for row in reader:
        Title.append(row[1])
        
print ""
print "-----------------------------"
print "Top 20 des recommandations personnalisées : "
print "-----------------------------"
for mov in Movies_ord[-20:] :
    print Title[mov]
    
# Comparaison avec des recommandations basées uniquement sur les notes moyennes

# Ordonne les films par ordre croissant de notes moyennes
Movies_ord_mean =  Y2_mean.argsort() 
Rates_ord_mean = Y2_mean[Y2_mean.argsort()]
Already_viewed_mean = R2[-1,Y2_mean.argsort()]

print ""
print "-----------------------------"
print "Top 20 des recommandations (basées sur les notes moyennes) : "
print "-----------------------------"
for mov in Movies_ord_mean[-20:] :
    print Title[mov]
```

### Quelques remarques sur les recommandations du nouvel utilisateur :

  * Il est étonnant que ce film figure aussi parmi les recommandations basées sur les notes moyennes
  * On peut comprendre l'origine du problème en extrayant le nombre d'utilisateurs ayant noté ce film :

* Un seul utilisateur parmi les 944 a noté ce film, et lui a attribué la note 5.0, ce qui place la note moyenne de ce film au maximum !
  * Suggestion pour contourner ce genre de soucis : il faudrait nettoyer le dataset pour ne garder que les films qui ont été notés par plus de (par exemple) 10 utilisateurs, afin d'éviter d'avoir des notes moyennes délirantes
* Autre souci : les recommandations "personnalisées" ne sont pas très différentes des recommandations basées uniquement sur les notes moyennes
  * Intuitivement, on peut incrimer l'utilisation d'un paramètre de régularisation identique pour Theta (critères utilisateurs) et X (critères films) : il se peut que les Theta soient trop contraints, empêchant les "goûts" des utilisateurs d'être correctement reflétés par le modèle
  * Dans la suite, on va construire un modèle similaire mais avec deux paramètres de régularisation distincts pour Theta et X

## Modification du modèle : définition de deux paramètres de régularisation indépendants pour Theta (rg1) et X (rg2)
* On définit deux nouvelles fonctions Cost_2reg et GradientDescent_2reg, respectivement similaires à Cost et GradientDescent
* On remplace l'argument rg des fonctions Cost et GradientDescent par deux arguments :
  * rg1 : paramètre de régularisation des Theta
  * rg 2 : paramètre de régularisation des X

```python
# Fonction de coût
def Cost_2reg(y, r, theta, x, rg1, rg2) :
    n_r = r[np.nonzero(r)].size
    error = (0.5/n_r)*np.einsum('ij,ji->', r*(np.dot(theta,x.T)-y), (r*(np.dot(theta,x.T)-y)).T)
    return error + (0.5*rg1/n_r)*np.einsum('ij,ji->', theta, theta.T) + (0.5*rg2/n_r)*np.einsum('ij,ji->', x, x.T)

# Méthode du gradient
def GradientDescent_2reg(y, r, theta, x, al, rg1, rg2, eps, max_it):
    n_r = r[np.nonzero(r)].size
    j = Cost_2reg(y, r, theta, x, rg1, rg2)
    j_hist = [j]
    j_old = j + 10*eps
    iter_n=0
    
    while abs(j-j_old) > eps :
        j_old = j
        Grad_X = (1.0/n_r)*np.dot((r*(np.dot(theta,x.T)-y)).T,theta) + (rg2/n_r)*x
        Grad_Theta = (1.0/n_r)*np.dot(r*(np.dot(theta,x.T)-y),x) + (rg1/n_r)*theta
        x = x - alpha*Grad_X
        theta = theta - alpha*Grad_Theta
        j = Cost_2reg(y, r, theta, x, rg1, rg2)
        j_hist.append(j)
        iter_n=iter_n+1
        if iter_n==max_it :
            print "max_iter at limit"
            break
    
    plt.plot(j_hist)
    plt.ylabel("Fonction de cout")
    plt.xlabel("nb iterations")
    plt.show()
    print "n_iter = " + repr(iter_n)
    print "coût à la convergence :" + repr(j_hist[-1])
    return [x,theta,iter_n,j_hist[-1]]
```

## Choix des paramètres de régularisation reg1 (Theta) et reg2 (X)
* On fixe reg1 à 0 (pas de régularisation des Theta pour l'instant)
* Pour différentes valeurs de reg2 :
  * On minimise la fonction de coût du train set
  * On évalue le coût sur le test set

```python
# Choix du paramètre de régularisation pour X

# Nombre de critères
n_features = 40
# Paramètre de régularisation pour Theta (on ne régularise pas Theta pour l'instant)
reg1 = 0.0
# Paramètre de régularisation pour X
reg2_values = [20.0,40.0,60.0,80.0,100.0,120.0,140.0,160.0]
# Learning rate
alpha = 300.0
# Condition de convergence
epsilon = 0.001
# Nombre maximal d'itération
max_iter = 200

iter_nb = []
J_train = []
J_test = []

for reg2 in reg2_values :
    X = np.random.randn(n_movies,n_features)
    Theta = np.random.randn(n_users,n_features)
    print "reg2 = " + repr(reg2)
    [X, Theta, a, b] = GradientDescent_2reg(Y_norm, R, Theta, X, alpha, reg1, reg2, epsilon, max_iter) # Entrainement training set avec régularisation
    iter_nb.append(a)
    J_train.append(Cost_2reg(Y_norm, R, Theta, X, 0.0, 0.0)) # Cout (sans les paramètres de régularisation)
    J_test.append(Cost_2reg(Y_test_norm, R_test, Theta, X, 0.0, 0.0)) # Evaluation du coût sur le test set (sans régularisation)
    print "-------------------------"
```

* On affiche J_train (en bleu) et J_test(en rouge) en fonction du paramètre reg2

```python
plt.plot(reg2_values, J_train, "b", reg2_values, J_test, "r")
plt.ylabel("cout")
plt.xlabel("parametre de regularisation")
plt.show()

print "regularisation : " + repr(reg2_values)
print "-------------------------------------"
print "cout sur le train set : " + repr(J_train)
print "-------------------------------------"
print "cout sur le test set : " + repr(J_test)
```

* Avec reg2=80.0, on cherche la meilleure valeur de reg1 (régularisation de Theta) :

## Méthode itérative pour déterminer les meilleurs paramètres de régularisation
* On fait une procédure itérative pour trouver les meilleures valeurs de reg1 et reg2 :

```python
# Méthode itérative pour déterminer les meilleurs paramètres de régularisation

# Nombre de critères
n_features = 40
# Paramètre de régularisation pour Theta (on ne régularise pas Theta pour l'instant)
reg1 = 2.5
# Paramètre de régularisation pour X
reg2 = 80.0
# Learning rate
alpha = 300.0
# Condition de convergence
epsilon = 0.001
# Nombre maximal d'itération
max_iter = 200

J_test_min = 0.4252
J_test_min_old = 1.0

while J_test_min_old - J_test_min > 0.001:
    J_test_min_old = J_test_min
    
    reg2_values = [reg2,reg2+10,reg2+20,reg2+30,reg2+40]
    if reg2-10.0>0.0 :
        reg2_values = np.append([reg2-10.0],reg2_values)
    if reg2-20.0>0.0 :
        reg2_values = np.append([reg2-20.0],reg2_values)
    if reg2-30.0>0.0 :
        reg2_values = np.append([reg2-30.0],reg2_values)
    if reg2-40.0>0.0 :
        reg2_values = np.append([reg2-40.0],reg2_values)
        
    print "----------------------------------------------------------------------------------------------------------------"
    print "Régularisation des Theta fixée à " + repr(reg1)    
    print "Régularisation des X : " + repr(reg2_values)
    print "-------------------------"
    iter_nb = []
    J_train = np.array([])
    J_test = np.array([])
    for reg2 in reg2_values :
        X = np.random.randn(n_movies,n_features)
        Theta = np.random.randn(n_users,n_features)
        print "reg2 = " + repr(reg2)
        [X, Theta, a, b] = GradientDescent_2reg(Y_norm, R, Theta, X, alpha, reg1, reg2, epsilon, max_iter) # Entrainement training set avec régularisation
        iter_nb.append(a)
        J_train = np.append(J_train, Cost_2reg(Y_norm, R, Theta, X, 0.0, 0.0)) # Cout (sans les paramètres de régularisation)
        J_test = np.append(J_test, Cost_2reg(Y_test_norm, R_test, Theta, X, 0.0, 0.0)) # Evaluation du coût sur le test set (sans régularisation)
        print "-------------------------"

    plt.plot(reg2_values, J_train, "b", reg2_values, J_test, "r")
    plt.ylabel("cout")
    plt.xlabel("parametre de regularisation des X")
    plt.show()

    reg2 = reg2_values[J_test.argmin()]
    J_test_min = J_test.min()
    print "-------------------------"
    print "J_test_min = " + repr(J_test_min) + " , pour reg2 = " + repr(reg2)
    
    if J_test_min_old - J_test_min < 0.001:
        break
        
    J_test_min_old = J_test_min
    
    reg1_values = [reg1,reg1+0.5,reg1+1.0,reg1+1.5,reg1+2.0]
    if reg1-0.5>0.0 :
        reg1_values = np.append([reg1-0.5],reg1_values)
    if reg1-1.0>0.0 :
        reg1_values = np.append([reg1-1.0],reg1_values)
    if reg1-1.5>0.0 :
        reg1_values = np.append([reg1-1.5],reg1_values)
    if reg1-2.0>0.0 :
        reg1_values = np.append([reg1-2.0],reg1_values)
        
    print "----------------------------------------------------------------------------------------------------------------"
    print "Régularisation des X fixée à " + repr(reg2)
    print "Régularisation des Theta : " + repr(reg1_values)
    print "-------------------------"
    iter_nb = []
    J_train = np.array([])
    J_test = np.array([])
    for reg1 in reg1_values :
        X = np.random.randn(n_movies,n_features)
        Theta = np.random.randn(n_users,n_features)
        print "reg1 = " + repr(reg1)
        [X, Theta, a, b] = GradientDescent_2reg(Y_norm, R, Theta, X, alpha, reg1, reg2, epsilon, max_iter) # Entrainement training set avec régularisation
        iter_nb.append(a)
        J_train = np.append(J_train, Cost_2reg(Y_norm, R, Theta, X, 0.0, 0.0)) # Cout (sans les paramètres de régularisation)
        J_test = np.append(J_test, Cost_2reg(Y_test_norm, R_test, Theta, X, 0.0, 0.0)) # Evaluation du coût sur le test set (sans régularisation)
        print "-------------------------"
    plt.plot(reg1_values, J_train, "b", reg1_values, J_test, "r")
    plt.ylabel("cout")
    plt.xlabel("parametre de regularisation des Theta")
    plt.show()

    reg1 = reg1_values[J_test.argmin()]
    J_test_min = J_test.min()
    print "-------------------------"
    print "J_test_min = " + repr(J_test_min) + " , pour reg1 = " + repr(reg1)
    
```

* Valeurs des paramètres de régularisation qui minimisent J_test : **reg1 = 5.0, reg2 = 40.0**
* On remarque que reg1 < 15.0 (valeur précédente de régularisation) et reg2 > 15.0, ce qui montre que le modèle précédent contraignait trop les Theta, et pas assez les X
* Pour les reg1 et reg2 retenus, J_train est de l'ordre de 0.33 alors que J_test est de l'ordre de 0.42. Cette différence de coût entre l'ensemble utilisé pour l'entraînement et l'ensemble test révèle qu'il subsiste un "overfitting" de l'ensemble d'entraînement. Une piste pour contrebalancer cet effet pourrait être d'utiliser un nombre de notes plus grand (ici on utilise l'ensemble MovieLens 100k contenant 100,000 notes, il faudrait envisager de tester le modèle avec l'ensemble MovieLens 1M qui contient 1 million de notes).
